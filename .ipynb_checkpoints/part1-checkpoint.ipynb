```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = []

for i in range(10):
    name = 'data/part2_pwm' + str(75 + i*10) +'.csv'
    tmp = pd.read_csv(name, index_col=0)
    tmp['resistance'] = - 47 * (tmp['scaled_voltage']  -5) / tmp['scaled_voltage'] 
    data.append(tmp)
tmp.head()
```

```python
all_bend = []
all_rest = []
plt.figure(figsize=(15,5))
for i,file in enumerate(data):
    
    plt.plot(file['bend_angle'], file['resistance'],'.', label = str(75 + i*10))
    
    all_bend.extend(file['bend_angle'])
    all_rest.extend(file['resistance'])
    
all_rest = np.array(all_rest).reshape(-1,1)
all_bend = np.array(all_bend).reshape(-1,1)

reg  =linear_model.LinearRegression()
reg.fit(all_bend, all_rest)
x = np.linspace(min(all_bend),max(all_bend))
plt.plot(x, reg.intercept_[0] + x*reg.coef_[0,0], label='Linear Regression', color ='black')
plt.legend()
plt.show()
print("intercept = ", reg.intercept_, "coef = ",reg.coef_)
```

```python
import pandas as pd
data = []
for i in range(10):
    name = 'data/part1_pwm' + str(75 + i*10) +'.csv'
    tmp = pd.read_csv(name, index_col=0)
    tmp['resistance'] = - 47 * (tmp['scaled_voltage']  -5) / tmp['scaled_voltage'] 
    tmp['bend_angle'] = 2 * (np.arctan((tmp['base_pos_y'] -tmp['tip_pos_y'])/(tmp['base_pos_x'] -tmp['tip_pos_x'])))
    data.append(tmp[10:100])
```

```python
all_bend = []
all_rest = []
plt.figure(figsize=(15,5))
for i,file in enumerate(data):
    
    plt.plot(file['bend_angle'], file['resistance'],'.', label = str(75 + i*10))
    
    all_bend.extend(file['bend_angle'])
    all_rest.extend(file['resistance'])
    
all_rest = np.array(all_rest).reshape(-1,1)
all_bend = np.array(all_bend).reshape(-1,1)

reg  =linear_model.LinearRegression()
reg.fit(all_bend, all_rest)
x = np.linspace(min(all_bend),max(all_bend))
plt.plot(x, reg.intercept_[0] + x*reg.coef_[0,0], label='Linear Regression', color ='black')
plt.xlabel('bend angle')
plt.ylabel('resistance')
plt.legend()
plt.show()
print("intercept = ", reg.intercept_, "coef = ",reg.coef_)
```
